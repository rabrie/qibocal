```python
import json
import numpy as np
```

```python
qubit = 2

with open(f"data/qw5q-qubit{qubit}-slow.json", "r") as file:
    data_slow = json.load(file)

with open(f"data/qw5q-qubit{qubit}-fast.json", "r") as file:
    data_fast = json.load(file)
```

```python
print(data_slow["nsequences"])
print(np.sum(data_slow["execution_times"]))
```

```python
print(data_fast["nsequences"])
print(np.sum(data_fast["sequence_creation_time"]))
print(np.sum(data_fast["execution_time"]))
```

```python
import qibo
from qibo import gates
from qibo.models import Circuit
qibo.set_backend("numpy")


def idx_to_probs(idx):
    c = Circuit(1)
    for i in idx:
        if i == 1:
            c.add(gates.RX(0, theta=np.pi / 2))
        elif i == 2:
            c.add(gates.RY(0, theta=np.pi / 2))
    c.add(gates.M(0))
    return list(np.around(c().probabilities(), decimals=5))
```

```python
simulation_probs = [idx_to_probs(idx) for idx in data_fast["sequences"]]

for x, y in zip(data_fast["hardware_probabilities"], simulation_probs):
    print(x, y)
```

```python
simulation_probs = [idx_to_probs(idx) for idx in data_slow["sequences"]]

for x, y in zip(data_slow["hardware_probabilities"], simulation_probs):
    print(x, y)
```
